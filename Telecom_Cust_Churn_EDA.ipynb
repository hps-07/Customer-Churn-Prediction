```python
telecom_data_copy = telecom_data.copy()  # Creating Copy for further convenience
```

```python
telecom_data_copy.TotalCharges = pd.to_numeric(telecom_data_copy.TotalCharges, errors='coerce')
telecom_data_copy.isnull().sum()
```

```python
telecom_data_copy.loc[telecom_data_copy['TotalCharges'].isnull()==True]
```

```python
telecom_data_copy.dropna(how='any',inplace=True)
#telecom_data_copy.fillna(0)
```

```python
print(telecom_data_copy['tenure'].max())
```

```python
labels = ["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
telecom_data_copy['tenure_group'] = pd.cut(telecom_data_copy.tenure,range(1,80,12),right=False,labels=labels) 
```

```python
telecom_data_copy['tenure_group'].value_counts()
```

```python
telecom_data_copy.drop(columns=['customerID','tenure'],axis=1,inplace=True)
telecom_data_copy.head()
```

Data Exploration : Univariate Analysis

```python
for i,predictor in enumerate(telecom_data_copy.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telecom_data_copy,x=predictor,hue='Churn')
```

```python
telecom_data_copy['Churn'] = np.where(telecom_data_copy.Churn == 'Yes',1,0)
```

```python
telecom_data_copy.head()
```

```python
telecom_data_copy_dummies = pd.get_dummies(telecom_data_copy).astype(int)
telecom_data_copy_dummies.head()
```

```python
sns.lmplot(data=telecom_data_copy_dummies,x='MonthlyCharges',y='TotalCharges',fit_reg=False)
```

```python
Monthly_charges = sns.kdeplot(telecom_data_copy_dummies.MonthlyCharges[(telecom_data_copy_dummies["Churn"]==0)],color='Red',fill=True)
Monthly_charges = sns.kdeplot(telecom_data_copy_dummies.MonthlyCharges[(telecom_data_copy_dummies["Churn"]==1)],ax=Monthly_charges,color='Blue',fill=True)
Monthly_charges.legend(["No Churn","Churn"],loc="upper right")
Monthly_charges.set_xlabel('Monthly Charges')
Monthly_charges.set_ylabel('Density')
Monthly_charges.set_title('Monthly Charges by Churn')
```

```python
Total_charges = sns.kdeplot(telecom_data_copy_dummies.TotalCharges[(telecom_data_copy_dummies["Churn"]==0)],color='Red',fill=True)
Total_charges = sns.kdeplot(telecom_data_copy_dummies.TotalCharges[(telecom_data_copy_dummies["Churn"]==1)],ax=Total_charges,color='Blue',fill=True)
Total_charges.legend(["No Churn","Churn"],loc="upper right")
Total_charges.set_xlabel('Total Charges')
Total_charges.set_ylabel('Density')
Total_charges.set_title('Total Charges by Churn')
```

```python
plt.figure(figsize=(20,8))
telecom_data_copy_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```

```python
def uniplot (df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"]=20
    plt.rcParams["axes.titlesize"]=22
    plt.rcParams["axes.titlepad"]=30
    
    temp = pd.Series(data=hue)
    fig,ax=plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data=df,x=col,order=df[col].value_counts().index,hue=hue,palette='bright')
    plt.show()
```

```python
uniplot(new_df1_target1,col="Partner",title="Distribution of Gender for Churned Customers",hue="gender")
```
